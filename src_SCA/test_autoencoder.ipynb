## Autoencoder

```python
# append the autoencoder training directory to path to import load_ae function
import sys
import os
sys.path.append(os.path.join(os.path.dirname("__file__"), "./train_ae_template/"))
from ae.autoencoder import load_ae
# import matplotlib for plotting
import matplotlib.pyplot as plt
# import cv2 to read images
import cv2
```

```python
# setup model path and test pictures path
model_path = r".\models\ae-32_1689619355_best.pkl"
test_pictures_path = r"..\data\dataset-mountain"
```

```python
# load the autoencoder model
ae_model = load_ae(model_path)
```

```python
# display one random image of the test picture dataset
all_images = os.listdir(test_pictures_path)
sample_image_path = test_pictures_path + "\\" + all_images[100]
sample_image = cv2.imread(sample_image_path)[:,:,::-1]
sample_image.shape
plt.imshow(sample_image)
plt.axis("off")
plt.savefig(r"..\data\sample_picture_training.jpg")
```

```python
# view latent space representation of image
encoded_image = ae_model.encode_from_raw_image(sample_image)
encoded_image, encoded_image.shape
```

```python
# reconstruct image
decoded_image = ae_model.decode(encoded_image)
decoded_image[0].shape
```

```python
# plot reconstructed image
plt.imshow(decoded_image[0][:,:,::-1])
plt.axis("off")
plt.savefig(r"..\data\reconstructed_picture.jpg")
```
